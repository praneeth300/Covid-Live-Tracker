To get the daily information about the cases of Covid-19.Related to the different features such as Active Cases,Deaths,New Recovered,New Cases all about informaton we can track by this process with help of covid live tracker

```python
#import the libraris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Get the data
data=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
data
```

```python
df=data.iloc[:-1,1:]
```

```python
df.head()
```

```python
total_states=np.arange(len(df['state_name']))
total_states
```

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(10,10),dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states,df['active'],align='center',alpha=0.5,color=(0,0,1),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['active'])+100)
plt.xlabel('Today Active cases in INDIA')
plt.title('Active Cases')
plt.show()
```

```python
figure(num=None,figsize=(10,10),dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states,df['positive'],alpha=0.8,color=(1,0,0),edgecolor=(0.1,0.2,0.3))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['positive'])+100)
plt.xlabel('Positive Cases Recorded in INDIA',fontsize=17)
plt.title('Positive Cases',fontsize=30)
plt.show()
```

```python
figure(num=None,figsize=(10,10),dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states,df['death'],alpha=0.8,color=(1,0,0),edgecolor=(0.1,0.2,0.3))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death'])+100)
plt.xlabel('Total Deaths INDIA',fontsize=17)
plt.title('Deaths',fontsize=30)
plt.show()
```

```python
df=df.set_index('state_name',drop=True)
```

```python
df.plot.barh(stacked=True,figsize=(10,10))
```

```python
df.head()
```

```python
df_1=df.iloc[:,:4]
df_1.plot.barh(color={"active":"red","positive":"blue","cured":"green","death":"black"},figsize=(10,20))
```

```python
figure(figsize=(9,9))
plt.barh(df['cured'].sort_values().index[-10:],df['cured'].sort_values().values[-10:])
for index,value in enumerate(df['cured'].sort_values().values[-10:]):
    plt.text(value,index,str(value))
plt.title('Top 10 States (Cured)INIDA',fontsize=20)
plt.xlabel('Cured Cases')
plt.show()
```

```python
figure(figsize=(9,9))
plt.barh(df['death'].sort_values().index[-10:],df['death'].sort_values().values[-10:])
for index,value in enumerate(df['death'].sort_values().values[-10:]):
    plt.text(value,index,str(value))
plt.title('Top 10 States (Deaths)INIDA',fontsize=20)
plt.xlabel('Death Range')
plt.show()
```

```python
figure(figsize=(9,9))
plt.barh(df['new_positive'].sort_values().index[-10:],df['new_positive'].sort_values().values[-10:])
for index,value in enumerate(df['new_positive'].sort_values().values[-10:]):
    plt.text(value,index,str(value))
plt.title('Top 10 States (New Postive)INIDA',fontsize=20)
plt.xlabel('Positive Cases')
plt.show()
```
